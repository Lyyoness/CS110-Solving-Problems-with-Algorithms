```python
import matplotlib.pyplot as plt
from scipy.stats import powerlaw, lognorm
import numpy as np

fig, ax = plt.subplots(1,1)

#power law
a = 1.66
rv = powerlaw(a)
x = np.linspace(powerlaw.ppf(0.01, a),
                powerlaw.ppf(0.99, a), 100)
ax.plot(x, rv.pdf(x), 'k-')

#log norm
s = 0.954
uv = lognorm(s)
y = np.linspace(lognorm.ppf(0.01, s),
                lognorm.ppf(0.99, s), 100)
ax.plot(x, uv.pdf(y),'r-', alpha=0.6)

#exponential
i = np.linspace(0,1,100)
j = np.exp(x)
ax.plot(i,j)
# ax.plot(i, -np.exp(-x))

#trig
t = np.arange(0.0, 1.0, 0.01)
ax.plot(t, np.sin(2*2*np.pi*t))

plt.show()
```
